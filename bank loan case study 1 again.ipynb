```python
import matplotlib.pyplot as plt

# List of columns to plot box plots and find outliers
columns_to_plot = ['AMT_ANNUITY', 'AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_GOODS_PRICE', 'DAYS_EMPLOYED', 'CNT_CHILDREN', 'DAYS_BIRTH']

# Create subplots for each column
fig, axes = plt.subplots(nrows=1, ncols=len(columns_to_plot), figsize=(12, 6))

# Iterate over columns and plot box plots
for i, column in enumerate(columns_to_plot):
    # Plot box plot
    axes[i].boxplot(df2[column])
    axes[i].set_title(column)

# Adjust spacing between subplots
plt.tight_layout()

# Find outliers
for column in columns_to_plot:
    # Calculate quartiles
    q1 = df2[column].quantile(0.25)
    q3 = df2[column].quantile(0.75)
    
    # Calculate interquartile range (IQR)
    iqr = q3 - q1
    
    # Calculate upper and lower bounds for outliers
    upper_bound = q3 + 1.5 * iqr
    lower_bound = q1 - 1.5 * iqr
    
    # Find outliers
    outliers = df2[(df2[column] > upper_bound) | (df2[column] < lower_bound)]
    print("Outliers in", column, ":", outliers)

# Show the box plots
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Group the data by 'NAME_EDUCATION_TYPE' and 'TARGET' columns and count the occurrences
grouped_data = df2.groupby(['NAME_EDUCATION_TYPE', 'TARGET']).size().unstack()

# Calculate the percentage of defaulters for each education type
grouped_data['Total'] = grouped_data.sum(axis=1)
grouped_data['Defaulter_Percentage'] = (grouped_data[0] / grouped_data['Total']) * 100

# Plot the percentage of defaulters
x = np.arange(len(grouped_data.index))

fig, ax = plt.subplots(figsize=(10, 6))
defaulter_percentage_bars = ax.bar(x, grouped_data['Defaulter_Percentage'], width=0.35)

# Add labels, title, and axis ticks
ax.set_xlabel('Education Type')
ax.set_ylabel('Percentage of Defaulters')
ax.set_title('Percentage of Defaulters by Education Type')
ax.set_xticks(x)
ax.set_xticklabels(grouped_data.index, rotation=45, ha='right')

# Display the values as percentages on top of the bars
for bar in defaulter_percentage_bars:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}%', xy=(bar.get_x() + bar.get_width() / 2, height), xytext=(0, 3),
                textcoords='offset points', ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```
